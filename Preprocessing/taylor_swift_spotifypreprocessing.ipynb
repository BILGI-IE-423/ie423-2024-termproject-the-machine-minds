```python
import pandas as pd
import matplotlib.pyplot as plt

# Let's get the first 5 data from the dataset
data = {
    'Track_Number': [1, 2, 3, 4, 5],
    'Valence': [0.685, 0.701, 0.305, 0.206, 0.520]
}

# Convert the dataset to a DataFrame
df = pd.DataFrame(data)

# Create a scatter plot showing the relationship between track number and valence
plt.scatter(df['Track_Number'], df['Valence'])
plt.xlabel('Track Number')
plt.ylabel('Valence')
plt.title('Relationship between Track Number and Valence')
plt.grid(True)  # Add grid
plt.show()
```
