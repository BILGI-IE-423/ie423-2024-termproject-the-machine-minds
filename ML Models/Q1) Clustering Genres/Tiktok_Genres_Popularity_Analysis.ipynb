```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tiktok = pd.read_csv('tiktok_predicted_genres_with_clusters.csv')
```

```python
popularity_by_genre = df_tiktok.groupby('predicted_genre')['track_pop'].mean().reset_index()
popularity_by_genre = popularity_by_genre.sort_values(by='track_pop', ascending=False)
```

```python
print(popularity_by_genre)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='track_pop', y='predicted_genre', data=popularity_by_genre, palette='viridis', hue='predicted_genre', dodge=False)
plt.legend([],[], frameon=False) 
plt.title('Mean Popularity by Genre')
plt.xlabel('Mean Popularity')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='track_pop', y='predicted_genre', data=df_tiktok, palette='viridis', hue='predicted_genre', dodge=False)
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend([],[], frameon=False)  # This will remove the legend
plt.title('Popularity Distribution by Genre')
plt.xlabel('Popularity')
plt.ylabel('Genre')
plt.show()
```
